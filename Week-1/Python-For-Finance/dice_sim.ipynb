# Dice Simulations for NumPy

```python
import numpy as np
import matplotlib.pyplot as plt
```

Let us now take in an input for the number of simulations to run.

```python
no_simulations = int(input("Please enter the number of simulations: "))
```

Now we create the data set for the simulations using the random method.

```python
arr = np.random.choice([1, 2, 3, 4, 5, 6], size = (no_simulations))
```

Now once we have our data set we will use matplotlib to plot the data into a histogram.

```python
plt.hist(arr)
plt.show()
```
